 Load the Dataset

```python
import pandas as pd

file_path = 'Job_Placement_Data.csv'
data = pd.read_csv(file_path)


data.head()
```

Bar Chart (Placements vs. Non-Placements per Degree)

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='undergrad_degree', hue='status', palette='Set2')
plt.title('Placements vs. Non-Placements per Undergraduate Degree')
plt.xlabel('Undergraduate Degree')
plt.ylabel('Count')
plt.legend(title='Status')
plt.show()
```

Pie Chart (Proportion of Placements by Degree)

```python

placements_by_degree = data[data['status'] == 'Placed']['undergrad_degree'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(placements_by_degree, labels=placements_by_degree.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Proportion of Placements by Undergraduate Degree')
plt.show()
```

Finding the Degree with the Most Placements

```python

placement_by_degree = data[data['status'] == 'Placed'].groupby('undergrad_degree')['status'].count()


most_placed_degree = placement_by_degree.idxmax()
placement_count = placement_by_degree.max()

print(f"The undergraduate degree with the most placements is '{most_placed_degree}' with {placement_count} placements.")
```
